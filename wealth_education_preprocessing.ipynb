## Research questions

**Q1.** How has the relationship between national wealth per capita and educational development changed over time?

**Q2.** Does wealth composition matter as much as wealth level? Do countries with higher human and produced capital shares achieve stronger educational outcomes than countries with greater dependence on natural capital?

**Q3.** Which countries overperform or underperform educationally relative to their wealth level, and are those patterns regionally clustered?

**Q4.** Among countries with similar wealth levels, do higher education expenditure and lower inequality correspond to stronger educational outcomes?

```python
import pandas as pd
import numpy as np
from pathlib import Path

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 200)

INPUT_CSV = Path("C:\\Users\\Samrat\\Downloads\\Samrat Ganguly K25131454 Data Viz Project\\raw\\export_data_q_analysis_ready.csv")

OUTPUT_DIR = Path("C:\\Users\\Samrat\\Downloads\\Samrat Ganguly K25131454 Data Viz Project\\outputs")
OUTPUT_DIR.mkdir(exist_ok=True)

INPUT_CSV, OUTPUT_DIR
```

```python
# Load master panel
df = pd.read_csv(INPUT_CSV)

print("Shape:", df.shape)
print("Columns:")
print(df.columns.tolist())

df.head()
```

## Patch the one metadata issue

The master panel is kept as the source-of-truth table.  
The only direct patch applied here is filling the missing income-group value for Ethiopia.

```python
# Patch Ethiopia's missing income group
eth_mask = (df["iso3"] == "ETH") & (df["income_group"].isna())
df.loc[eth_mask, "income_group"] = "Low income"

print("Remaining missing income_group rows:", df["income_group"].isna().sum())
df[df["iso3"] == "ETH"][["iso3", "country_display", "income_group"]].drop_duplicates()
```

## Quick validation checks

The master panel should still be one row per country-year.

```python
dup_count = df.duplicated(subset=["iso3", "year"]).sum()
print("Duplicate country-year rows:", dup_count)
print("Year range:", int(df["year"].min()), "to", int(df["year"].max()))
print("Number of countries:", df["iso3"].nunique())

quality_cols = [
    "gini_index",
    "school_life_expectancy",
    "tertiary_enrolment_rate",
    "lower_secondary_completion_rate",
    "education_expenditure_pct_gdp",
]
(df[quality_cols].notna().mean().sort_values(ascending=False).rename("non_null_share").to_frame())
```

##  `q_analysis_core`

- remove year 2003 because the education extract has a structural gap there
- keep rows where at least one of the two strongest education variables is present:
  - `school_life_expectancy`
  - `tertiary_enrolment_rate`

All original columns are retained, including the share variables.

```python
analysis_core = df[
    (df["year"] != 2003) &
    (
        df["school_life_expectancy"].notna() |
        df["tertiary_enrolment_rate"].notna()
    )
].copy()

print("q_analysis_core shape:", analysis_core.shape)
print("Contains 2003:", (analysis_core["year"] == 2003).any())
print("All rows have at least one core education variable:",
      ((analysis_core["school_life_expectancy"].notna()) | (analysis_core["tertiary_enrolment_rate"].notna())).all())

analysis_core.head()
```

## Create `q_q4_spend`

This is the policy subset used for the education-spending part of Q4.

```python
q4_spend = analysis_core[
    analysis_core["education_expenditure_pct_gdp"].notna()
].copy()

print("q_q4_spend shape:", q4_spend.shape)
q4_spend.head()
```

## Create `q_q4_gini`

This is the stricter policy subset used for the inequality + education-spending part of Q4.

```python
q4_gini = q4_spend[
    q4_spend["gini_index"].notna()
].copy()

print("q_q4_gini shape:", q4_gini.shape)
q4_gini.head()
```

## Create `q_country_summary`

This table is used for country-level comparisons and the overperformer / underperformer analysis (Q3).

```python
def count_notna(s):
    return s.notna().sum()

country_summary = (
    analysis_core
    .groupby(["iso3", "country_display", "region", "income_group"], dropna=False)
    .agg(
        avg_log_wealth_pc=("log_wealth_pc", "mean"),
        avg_wealth_pc=("wealth_pc", "mean"),
        avg_human_capital_share=("human_capital_share", "mean"),
        avg_natural_capital_share=("natural_capital_share", "mean"),
        avg_produced_capital_share=("produced_capital_share", "mean"),
        avg_school_life_expectancy=("school_life_expectancy", "mean"),
        avg_tertiary_enrolment_rate=("tertiary_enrolment_rate", "mean"),
        avg_lower_secondary_completion_rate=("lower_secondary_completion_rate", "mean"),
        avg_education_expenditure_pct_gdp=("education_expenditure_pct_gdp", "mean"),
        avg_gini_index=("gini_index", "mean"),
        n_years_total=("year", "size"),
        n_school_years=("school_life_expectancy", count_notna),
        n_tertiary_years=("tertiary_enrolment_rate", count_notna),
        n_spend_years=("education_expenditure_pct_gdp", count_notna),
        n_gini_years=("gini_index", count_notna),
    )
    .reset_index()
)

print("q_country_summary shape:", country_summary.shape)
country_summary.head()
```

## Create `q_year_summary`

This table supports Q1 by showing how the central tendency of the relationship changes over time.

Median is used for skewed variables rather than mean.

```python
year_summary = (
    analysis_core
    .groupby("year", dropna=False)
    .agg(
        median_log_wealth_pc=("log_wealth_pc", "median"),
        median_school_life_expectancy=("school_life_expectancy", "median"),
        median_tertiary_enrolment_rate=("tertiary_enrolment_rate", "median"),
        median_human_capital_share=("human_capital_share", "median"),
        median_natural_capital_share=("natural_capital_share", "median"),
        median_produced_capital_share=("produced_capital_share", "median"),
        median_education_expenditure_pct_gdp=("education_expenditure_pct_gdp", "median"),
        median_gini_index=("gini_index", "median"),
        n_countries=("iso3", "nunique"),
        n_school_obs=("school_life_expectancy", count_notna),
        n_tertiary_obs=("tertiary_enrolment_rate", count_notna),
        n_spend_obs=("education_expenditure_pct_gdp", count_notna),
        n_gini_obs=("gini_index", count_notna),
    )
    .reset_index()
)

print("q_year_summary shape:", year_summary.shape)
year_summary.head()
```

## Create `q_year_region_summary`

This optional table supports region-by-time comparisons.

```python
year_region_summary = (
    analysis_core
    .groupby(["year", "region"], dropna=False)
    .agg(
        median_log_wealth_pc=("log_wealth_pc", "median"),
        median_school_life_expectancy=("school_life_expectancy", "median"),
        median_tertiary_enrolment_rate=("tertiary_enrolment_rate", "median"),
        median_human_capital_share=("human_capital_share", "median"),
        median_natural_capital_share=("natural_capital_share", "median"),
        median_produced_capital_share=("produced_capital_share", "median"),
        n_countries=("iso3", "nunique"),
    )
    .reset_index()
)

print("q_year_region_summary shape:", year_region_summary.shape)
year_region_summary.head()
```

## Data-quality helper exports

These are useful for checking missingness by variable and by year before building Power BI pages.

```python
coverage_by_variable = pd.DataFrame({
    "column": df.columns,
    "non_null_count": [df[c].notna().sum() for c in df.columns],
    "non_null_share": [df[c].notna().mean() for c in df.columns],
    "dtype": [str(df[c].dtype) for c in df.columns],
}).sort_values(["non_null_share", "column"], ascending=[False, True])

coverage_by_year = (
    df.groupby("year")[[
        "school_life_expectancy",
        "tertiary_enrolment_rate",
        "lower_secondary_completion_rate",
        "education_expenditure_pct_gdp",
        "gini_index",
    ]]
    .apply(lambda g: g.notna().mean())
    .reset_index()
)

coverage_by_variable.head(10), coverage_by_year.head()
```

## Export all derived CSVs

```python
output_files = {
    "q_analysis_core": OUTPUT_DIR / "q_analysis_core.csv",
    "q_q4_spend": OUTPUT_DIR / "q_q4_spend.csv",
    "q_q4_gini": OUTPUT_DIR / "q_q4_gini.csv",
    "q_country_summary": OUTPUT_DIR / "q_country_summary.csv",
    "q_year_summary": OUTPUT_DIR / "q_year_summary.csv",
    "q_year_region_summary": OUTPUT_DIR / "q_year_region_summary.csv",
    "q_coverage_by_variable": OUTPUT_DIR / "q_coverage_by_variable.csv",
    "q_coverage_by_year": OUTPUT_DIR / "q_coverage_by_year.csv",
}

analysis_core.to_csv(output_files["q_analysis_core"], index=False)
q4_spend.to_csv(output_files["q_q4_spend"], index=False)
q4_gini.to_csv(output_files["q_q4_gini"], index=False)
country_summary.to_csv(output_files["q_country_summary"], index=False)
year_summary.to_csv(output_files["q_year_summary"], index=False)
year_region_summary.to_csv(output_files["q_year_region_summary"], index=False)
coverage_by_variable.to_csv(output_files["q_coverage_by_variable"], index=False)
coverage_by_year.to_csv(output_files["q_coverage_by_year"], index=False)

for name, path in output_files.items():
    print(f"{name}: {path}")
```
